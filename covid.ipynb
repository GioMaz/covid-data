```python
import pandas as pd

vax = pd.read_csv('https://raw.githubusercontent.com/italia/covid19-opendata-vaccini/master/dati/somministrazioni-vaccini-latest.csv')
vax.data_somministrazione = pd.to_datetime(vax.data_somministrazione)
andamento = pd.read_json('https://raw.githubusercontent.com/pcm-dpc/COVID-19/master/dati-json/dpc-covid19-ita-andamento-nazionale.json')
andamento.data = pd.to_datetime(andamento.data)
```

```python
vax.tail()
```

```python
andamento.tail()
```

```python
vax.prima_dose.plot()
vax.seconda_dose.plot()
vax.dose_addizionale_booster.plot()
```

```python
andamento.totale_positivi.plot()
andamento.terapia_intensiva.plot()
```

```python
plot_prima_dose = vax.plot(x='data_somministrazione', y='prima_dose')
andamento.plot(x='data', y='terapia_intensiva', ax=plot_prima_dose)
```

```python
plot_seconda_dose = vax.plot(x='data_somministrazione', y='seconda_dose')
andamento.plot(x='data', y='terapia_intensiva', ax=plot_seconda_dose)
```

```python
plot_terza_dose = vax.plot(x='data_somministrazione', y='dose_addizionale_booster')
andamento.plot(x='data', y='terapia_intensiva', ax=plot_terza_dose)
```

```python
plot_dosi = vax.plot(x='data_somministrazione', y=['prima_dose', 'seconda_dose', 'dose_addizionale_booster'])
andamento.plot(x='data', y='terapia_intensiva', ax=plot_dosi)
```
